```python
# informer_binary_rain.py
# Informer-style ProbSparse attention (TensorFlow/Keras)
# Author: ChatGPT (adapted to your dataset/pipeline)

import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Dense, LayerNormalization, Dropout, GlobalAveragePooling1D,
    Add, Reshape, Flatten
)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt

# ---------------------------
# 1) Data loading & preprocessing (same year-split as you used)
# ---------------------------
def load_and_prepare_data(csv_paths):
    # csv_paths: dict with keys 2019,2020,2021,2022 -> path strings
    frames = []
    for year, path in csv_paths.items():
        df = pd.read_csv(path).assign(YEAR=int(year))
        frames.append(df)
    full_df = pd.concat(frames, ignore_index=True)

    # Basic required columns
    required_cols = ['YEAR', 'MO', 'DY', 'PRECTOTCORR']
    for col in required_cols:
        if col not in full_df.columns:
            raise ValueError(f"Missing {col} in combined dataframe")

    # numeric conversion & drop
    for c in required_cols:
        full_df[c] = pd.to_numeric(full_df[c], errors='coerce')
    full_df.dropna(subset=required_cols, inplace=True)

    # datetime features
    full_df['datetime'] = pd.to_datetime(dict(year=full_df['YEAR'],
                                              month=full_df['MO'],
                                              day=full_df['DY']))
    full_df['DOY'] = full_df['datetime'].dt.dayofyear
    full_df['sin_DOY'] = np.sin(2 * np.pi * full_df['DOY'] / 365.25)
    full_df['cos_DOY'] = np.cos(2 * np.pi * full_df['DOY'] / 365.25)

    # lags
    for lag in [1, 3, 7]:
        full_df[f'PRECTOTCORR_lag{lag}'] = full_df['PRECTOTCORR'].shift(lag)

    # SLT as earlier
    full_df['SLT'] = full_df['TS'] + (0.0065 * 505.5)

    full_df.dropna(inplace=True)

    features = [
        'SLT','SLP','T2M','TS','T2M_MAX','T2M_MIN',
        'RH2M','WS10M_MAX','WS10M_MIN',
        'sin_DOY','cos_DOY',
        'PRECTOTCORR_lag1','PRECTOTCORR_lag3','PRECTOTCORR_lag7'
    ]
    full_df['PRECTOTCORR_binary'] = (full_df['PRECTOTCORR'] > 0).astype(int)

    # split
    train_df = full_df[full_df['YEAR'].isin([2019,2020,2021])].copy()
    test_df = full_df[full_df['YEAR'] == 2022].copy()

    # safety checks
    if train_df.empty or test_df.empty:
        raise ValueError("Train/test split empty - check CSV paths / years")

    print("Train samples:", len(train_df), "Test samples:", len(test_df))
    print("Train class dist:\n", train_df['PRECTOTCORR_binary'].value_counts(normalize=True))
    print("Test class dist:\n", test_df['PRECTOTCORR_binary'].value_counts(normalize=True))

    return train_df, test_df, features, 'PRECTOTCORR_binary'

# ---------------------------
# 2) Scaling + sequence creation
# ---------------------------
def preprocess_and_sequences(train_df, test_df, features, target, seq_len, scaler=None):
    # fit scaler on training features (per time-step features)
    if scaler is None:
        scaler = MinMaxScaler()
        scaler.fit(train_df[features])
    train_scaled = scaler.transform(train_df[features])
    test_scaled = scaler.transform(test_df[features])

    # create DataFrames with same index as originals for temporal integrity
    train_proc = pd.DataFrame(train_scaled, columns=features, index=train_df.index)
    train_proc[target] = train_df[target].values

    test_proc = pd.DataFrame(test_scaled, columns=features, index=test_df.index)
    test_proc[target] = test_df[target].values
    test_proc['DOY'] = test_df['DOY'].values

    # create sequences (rolling window, preserving order)
    def create_seq_from_df(df):
        X, y = [], []
        # we must use continuous index order
        for i in range(len(df) - seq_len):
            X.append(df[features].iloc[i:i+seq_len].values)
            y.append(df[target].iloc[i+seq_len])
        return np.array(X), np.array(y)

    X_train, y_train = create_seq_from_df(train_proc)
    X_test, y_test = create_seq_from_df(test_proc)

    print("X_train shape:", X_train.shape, "X_test shape:", X_test.shape)
    return X_train, y_train, X_test, y_test, scaler

# ---------------------------
# 3) ProbSparse-ish Multi-Head Attention layer
#   - For each query, keep top_k keys by score (approximate)
#   - Implemented in TF ops; not fully optimized but works for moderate sizes.
# ---------------------------
class ProbSparseMultiHeadAttention(tf.keras.layers.Layer):
    def __init__(self, d_model, num_heads, top_k_ratio=0.2, dropout=0.1, **kwargs):
        super().__init__(**kwargs)
        assert d_model % num_heads == 0
        self.d_model = d_model
        self.num_heads = num_heads
        self.depth = d_model // num_heads
        self.top_k_ratio = top_k_ratio
        self.dropout = dropout

        self.wq = Dense(d_model)
        self.wk = Dense(d_model)
        self.wv = Dense(d_model)
        self.out_dense = Dense(d_model)
        self.attn_dropout = tf.keras.layers.Dropout(dropout)

    def split_heads(self, x):
        # x shape: (batch, seq_len, d_model)
        batch = tf.shape(x)[0]
        seq_len = tf.shape(x)[1]
        x = tf.reshape(x, (batch, seq_len, self.num_heads, self.depth))
        return tf.transpose(x, perm=[0,2,1,3])  # (batch, heads, seq_len, depth)

    def call(self, q_input, k_input, v_input, training=False):
        q = self.wq(q_input)
        k = self.wk(k_input)
        v = self.wv(v_input)

        q = self.split_heads(q)
        k = self.split_heads(k)
        v = self.split_heads(v)

        # scaled dot-product
        dk = tf.cast(self.depth, tf.float32)
        logits = tf.matmul(q, k, transpose_b=True) / tf.math.sqrt(dk)  # (batch, heads, q_len, k_len)

        # Approximate ProbSparse: for each query, keep top_k keys (by max logits)
        k_len = tf.shape(logits)[-1]
        # compute top_k per head per batch
        top_k = tf.maximum(1, tf.cast(tf.math.ceil(tf.cast(k_len, tf.float32) * self.top_k_ratio), tf.int32))

        # we need to mask out everything except top_k for each query
        # find top_k indices for each (batch, head, q_pos)
        # tf.nn.top_k operates on last dim
        values, indices = tf.nn.top_k(logits, k=top_k, sorted=False)  # shapes (batch,heads,q_len,top_k)

        # Create a mask with very low scores where not selected
        # Build a sparse mask using scatter
        batch = tf.shape(logits)[0]
        heads = tf.shape(logits)[1]
        q_len = tf.shape(logits)[2]

        # We'll construct a mask by setting default -inf then scatter the kept values
        neg_inf = -1e9
        mask = tf.ones_like(logits) * neg_inf  # (batch,heads,q_len,k_len)

        # Scatter the selected logits into mask
        # Prepare indices for scatter_nd: need shape [N,4] for (batch, head, q_pos, k_idx)
        # We'll vectorize: build coords tensors
        b_idx = tf.range(batch)[:,None,None,None]    # (batch,1,1,1)
        h_idx = tf.range(heads)[None,:,None,None]   # (1,heads,1,1)
        q_idx = tf.range(q_len)[None,None,:,None]   # (1,1,q_len,1)

        b_idx = tf.broadcast_to(b_idx, tf.shape(indices))
        h_idx = tf.broadcast_to(h_idx, tf.shape(indices))
        q_idx = tf.broadcast_to(q_idx, tf.shape(indices))

        scatter_idx = tf.stack([tf.reshape(b_idx,[-1]),
                                tf.reshape(h_idx,[-1]),
                                tf.reshape(q_idx,[-1]),
                                tf.reshape(indices,[-1])], axis=1)  # (batch*heads*q_len*top_k,4)

        scatter_vals = tf.reshape(values, [-1])

        mask = tf.tensor_scatter_nd_update(mask, scatter_idx, scatter_vals)

        # softmax over last dim with masked logits
        attn_weights = tf.nn.softmax(mask, axis=-1)
        attn_weights = self.attn_dropout(attn_weights, training=training)

        attn_out = tf.matmul(attn_weights, v)  # (batch,heads,q_len,depth)
        attn_out = tf.transpose(attn_out, perm=[0,2,1,3])  # (batch,q_len,heads,depth)
        attn_out = tf.reshape(attn_out, (batch, q_len, self.d_model))
        out = self.out_dense(attn_out)
        return out

# ---------------------------
# 4) Informer-style encoder block
# ---------------------------
def informer_encoder_block(x, d_model, num_heads, ff_dim, top_k_ratio=0.2, dropout=0.1):
    attn_layer = ProbSparseMultiHeadAttention(d_model, num_heads, top_k_ratio=top_k_ratio, dropout=dropout)
    attn_out = attn_layer(x, x, x)
    x = Add()([x, attn_out])
    x = LayerNormalization()(x)

    # position-wise feed-forward
    ffn = Dense(ff_dim, activation='relu')(x)
    ffn = Dense(d_model)(ffn)
    ffn = Dropout(dropout)(ffn)
    x = Add()([x, ffn])
    x = LayerNormalization()(x)
    return x

# ---------------------------
# 5) Build Informer-like model (encoder-only for classification)
# ---------------------------
def build_informer_classifier(seq_len, n_features,
                              d_model=128, num_heads=4, ff_dim=256,
                              num_blocks=3, top_k_ratio=0.2, dropout=0.1):
    inp = Input(shape=(seq_len, n_features))
    # simple linear projection to d_model
    x = Dense(d_model)(inp)

    # optionally add positional encodings (sinusoidal)
    pos = positional_encoding_1d(seq_len, d_model)
    x = x + pos

    # stacked encoder blocks
    for _ in range(num_blocks):
        x = informer_encoder_block(x, d_model, num_heads, ff_dim, top_k_ratio=top_k_ratio, dropout=dropout)

    # pooling + classifier head
    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.2)(x)
    out = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inp, outputs=out)
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
        loss='binary_crossentropy',
        metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()]
    )
    model.summary()
    return model

# sinusoidal positional encoding (1D)
def positional_encoding_1d(seq_len, d_model):
    # returns tensor shape (1, seq_len, d_model)
    position = np.arange(seq_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    pe = np.zeros((seq_len, d_model))
    pe[:, 0::2] = np.sin(position * div_term)
    pe[:, 1::2] = np.cos(position * div_term)
    pe = pe[np.newaxis, ...].astype(np.float32)
    return tf.constant(pe)

# ---------------------------
# 6) Training + evaluation + plotting
# ---------------------------
def plot_results(history, y_test, y_pred_prob, y_pred):
    plt.figure(figsize=(14,8))
    plt.subplot(2,2,1)
    plt.plot(history.history['loss'], label='train_loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.legend(); plt.title('Loss')

    plt.subplot(2,2,2)
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
    plt.plot(fpr, tpr, label=f'AUC={auc(fpr,tpr):.3f}')
    plt.plot([0,1],[0,1],'k--'); plt.legend(); plt.title('ROC')

    plt.subplot(2,2,3)
    cm = confusion_matrix(y_test, y_pred)
    plt.imshow(cm, cmap='Blues'); plt.title('Confusion Matrix')
    plt.colorbar()

    plt.subplot(2,2,4)
    plt.plot(y_test[:200], 'o-', label='actual')
    plt.plot(y_pred[:200], 'x--', label='pred')
    plt.legend(); plt.title('First 200 samples')

    plt.tight_layout(); plt.show()

def train_and_evaluate_informer(csv_paths, seq_len=90, batch_size=16, epochs=120,
                                 d_model=128, num_heads=4, ff_dim=256, num_blocks=3,
                                 top_k_ratio=0.2, dropout=0.1):
    # 1) load
    train_df, test_df, features, target = load_and_prepare_data(csv_paths)

    # 2) preprocess & sequences
    X_train, y_train, X_test, y_test, scaler = preprocess_and_sequences(train_df, test_df, features, target, seq_len)

    # 3) compute class weights (avoid SMOTE for sequence data)
    classes = np.unique(y_train)
    class_weights = compute_class_weight('balanced', classes=classes, y=y_train)
    class_weight_dict = {int(c): w for c,w in zip(classes, class_weights)}
    print("Class weights:", class_weight_dict)

    # 4) build model
    model = build_informer_classifier(seq_len, len(features),
                                      d_model=d_model, num_heads=num_heads,
                                      ff_dim=ff_dim, num_blocks=num_blocks,
                                      top_k_ratio=top_k_ratio, dropout=dropout)

    # 5) callbacks
    callbacks = [
        EarlyStopping(patience=12, restore_best_weights=True, monitor='val_loss'),
        ReduceLROnPlateau(patience=4, factor=0.5, min_lr=1e-6, monitor='val_loss')
    ]

    # 6) fit
    history = model.fit(X_train, y_train,
                        validation_split=0.2,
                        epochs=epochs,
                        batch_size=batch_size,
                        callbacks=callbacks,
                        class_weight=class_weight_dict,
                        verbose=1)

    # 7) evaluate
    y_pred_prob = model.predict(X_test, batch_size=64).ravel()
    y_pred = (y_pred_prob > 0.5).astype(int)

    print("\n=== Test Set Evaluation ===")
    print("Accuracy: ", accuracy_score(y_test, y_pred))
    print("Precision:", precision_score(y_test, y_pred))
    print("Recall:   ", recall_score(y_test, y_pred))
    print("F1:       ", f1_score(y_test, y_pred))
    print("ROC-AUC:  ", roc_auc_score(y_test, y_pred_prob))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    plot_results(history, y_test, y_pred_prob, y_pred)
    return model, scaler, (X_test, y_test)

# ---------------------------
# 7) Example usage
# ---------------------------
if __name__ == "__main__":
    # point these to your CSVs (local paths)
    csv_paths = {
        2019: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2019.csv",
        2020: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2020.csv",
        2021: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2021.csv",
        2022: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2022.csv",
    }

    model, scaler, test_tuple = train_and_evaluate_informer(csv_paths,
                                                           seq_len=15,
                                                           batch_size=8,   # lower if OOM
                                                           epochs=120,
                                                           d_model=128,
                                                           num_heads=4,
                                                           ff_dim=256,
                                                           num_blocks=3,
                                                           top_k_ratio=0.25,
                                                           dropout=0.12)
```
