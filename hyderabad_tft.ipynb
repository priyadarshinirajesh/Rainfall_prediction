```python
# rainfall_models_adilabad.py
# Three pipelines (CNN-LSTM, Transformer, TabNet+Sequence) sharing the same preprocessing.
# Author: (adapted for your request)
# in this, i have added more lag features (lag14, lag30) - helps in focusing on large monthly seasonal cycle

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve, auc
)
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import *
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# -----------------------------
# Config (change as needed)
# -----------------------------
DATA_FILES = {
    2019: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2019.csv",
    2020: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2020.csv",
    2021: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2021.csv",
    2022: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2022.csv",
    2023: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2023.csv",
    2024: r"C:\Users\priya\rainfall_prediction\Rainfall_prediction\datasets\hyderabad_2024.csv",
}
TRAIN_YEARS = [2019, 2020, 2021]
TEST_YEAR = 2022
SEQ_LENGTH = 30      # sliding window length (you can increase to 30/60 to see effects)
BATCH_SIZE = 32
EPOCHS = 100
RANDOM_STATE = 42

# -----------------------------
# Utility: load + prepare dataset
# -----------------------------
def load_and_prepare_data(file_map=DATA_FILES):
    # load all csvs listed in file_map (keys = year)
    dfs = []
    for year in sorted(file_map.keys()):
        path = file_map[year]
        df = pd.read_csv(path).assign(YEAR=year)
        dfs.append(df)
    full_df = pd.concat(dfs, ignore_index=True)

    # required columns check (match your earlier code)
    required_cols = ['YEAR', 'MO', 'DY', 'PRECTOTCORR']
    missing = [c for c in required_cols if c not in full_df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

    # ensure numeric
    for c in required_cols:
        full_df[c] = pd.to_numeric(full_df[c], errors='coerce')
    full_df.dropna(subset=required_cols, inplace=True)

    # datetime and cyclic DOY features
    full_df['datetime'] = pd.to_datetime(dict(year=full_df['YEAR'], month=full_df['MO'], day=full_df['DY']))
    full_df['DOY'] = full_df['datetime'].dt.dayofyear
    full_df['sin_DOY'] = np.sin(2*np.pi*full_df['DOY']/365.25)
    full_df['cos_DOY'] = np.cos(2*np.pi*full_df['DOY']/365.25)

    # lags
    for lag in [1, 3, 7, 14, 30]:
        full_df[f'PRECTOTCORR_lag{lag}'] = full_df['PRECTOTCORR'].shift(lag)

    # sea level temp: hyderabad elev = 542m (your formula)
    full_df['SLT'] = full_df['TS'] + (0.0065 * 542)

    # drop any remaining NaNs
    full_df.dropna(inplace=True)

    # features and target (binary)
    features = ['SLT', 'SLP', 'T2M', 'TS', 'T2M_MAX', 'T2M_MIN',
            'RH2M', 'WS10M_MAX', 'WS10M_MIN',
            'sin_DOY', 'cos_DOY',
            'PRECTOTCORR_lag1', 'PRECTOTCORR_lag3', 'PRECTOTCORR_lag7',
            'PRECTOTCORR_lag14', 'PRECTOTCORR_lag30']

    full_df['Rain'] = (full_df['PRECTOTCORR'] > 0).astype(int)

    train_df = full_df[full_df['YEAR'].isin(TRAIN_YEARS)].copy()
    test_df = full_df[full_df['YEAR'] == TEST_YEAR].copy()

    # temporal leakage check
    if len(train_df) == 0 or len(test_df) == 0:
        raise ValueError("Train or test data empty. Check years and files.")
    if test_df['datetime'].min() <= train_df['datetime'].max():
        # Typically okay as long as test is separate year; still keep check
        pass

    print("TRAIN SIZE:", len(train_df), "TEST SIZE:", len(test_df))
    print("Train class distribution:\n", train_df['Rain'].value_counts(normalize=True))
    return train_df, test_df, features, 'Rain'

# -----------------------------
# Preprocess: scaling + SMOTE + sequences
# -----------------------------
def preprocess_and_balance(train_df, test_df, features, target):
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(train_df[features])
    X_test_scaled = scaler.transform(test_df[features])

    # SMOTE on training data (tabular)
    sm = SMOTE(random_state=RANDOM_STATE)
    X_res, y_res = sm.fit_resample(X_train_scaled, train_df[target].values)

    # create balanced train DataFrame (ordered by resample index -> shuffled)
    train_balanced = pd.DataFrame(X_res, columns=features)
    train_balanced[target] = y_res

    test_processed = pd.DataFrame(X_test_scaled, columns=features, index=test_df.index)
    test_processed[target] = test_df[target].values
    test_processed['DOY'] = test_df['DOY']

    print("After SMOTE balancing:", train_balanced[target].value_counts())
    return train_balanced, test_processed, scaler

def create_sequences(data_df, features, target, seq_length):
    X, y = [], []
    for i in range(len(data_df) - seq_length):
        X.append(data_df[features].iloc[i:i+seq_length].values)
        y.append(data_df[target].iloc[i+seq_length])
    X = np.array(X)
    y = np.array(y)
    return X, y

# -----------------------------
# Model A: CNN-LSTM
# -----------------------------
def build_cnn_lstm(seq_len, n_features):
    inp = Input(shape=(seq_len, n_features))
    x = Conv1D(64, kernel_size=3, padding='same', activation='relu')(inp)
    x = Conv1D(64, kernel_size=3, padding='same', activation='relu')(x)
    x = MaxPooling1D(pool_size=2)(x)
    x = LSTM(64)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.3)(x)
    out = Dense(1, activation='sigmoid')(x)
    model = Model(inp, out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()])
    model.summary()
    return model

# -----------------------------
# Model B: Transformer (simple encoder)
# -----------------------------
def build_transformer_binary(seq_len, n_features, d_model=128, heads=4, ff_dim=256):
    inp = Input(shape=(seq_len, n_features))
    # linear projection to d_model
    x = Dense(d_model)(inp)
    # learnable positional embedding
    pos_emb = tf.Variable(tf.random.normal([1, seq_len, d_model]), trainable=True)
    x = x + pos_emb

    for _ in range(2):  # two encoder blocks
        attn = MultiHeadAttention(num_heads=heads, key_dim=d_model//heads)(x, x)
        attn = Dropout(0.2)(attn)
        x = LayerNormalization()(x + attn)

        ffn = Dense(ff_dim, activation='relu')(x)
        ffn = Dense(d_model)(ffn)
        ffn = Dropout(0.2)(ffn)
        x = LayerNormalization()(x + ffn)

    x = GlobalAveragePooling1D()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    out = Dense(1, activation='sigmoid')(x)
    model = Model(inp, out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()])
    model.summary()
    return model

# -----------------------------
# Model C: Temporal Fusion Transformer (TFT)
# -----------------------------

def build_tft_model(seq_len, n_features, hidden_size=64, dropout=0.2):

    seq_inp = Input(shape=(seq_len, n_features))

    # 1. Variable Selection Network
    vsn = Dense(hidden_size, activation="relu")(seq_inp)
    vsn = Dense(n_features, activation="softmax")(vsn)
    selected = Multiply()([seq_inp, vsn])

    # 2. Local GRU encoder
    enc = GRU(hidden_size, return_sequences=True)(selected)

    # 3. Multi-head attention
    attn = MultiHeadAttention(num_heads=4, key_dim=hidden_size)(enc, enc)
    attn = Dropout(dropout)(attn)
    attn_out = LayerNormalization()(enc + attn)

    # 4. Feed-Forward Network with gating
    ffn = Dense(hidden_size*2, activation="relu")(attn_out)
    ffn = Dense(hidden_size)(ffn)
    gated = Multiply()([ffn, Dense(hidden_size, activation="sigmoid")(ffn)])
    out = LayerNormalization()(attn_out + gated)

    # 5. Decoder (Classification Head)
    x = GlobalAveragePooling1D()(out)
    x = Dense(64, activation="relu")(x)
    x = Dropout(dropout)(x)
    final_out = Dense(1, activation="sigmoid")(x)

    model = Model(seq_inp, final_out)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss="binary_crossentropy",
        metrics=[
            "accuracy",
            tf.keras.metrics.Precision(),
            tf.keras.metrics.Recall(),
            tf.keras.metrics.AUC()
        ]
    )

    model.summary()
    return model

def train_tft_pipeline(seq_len=SEQ_LENGTH):
    train_df, test_df, features, target = load_and_prepare_data()
    train_balanced, test_processed, scaler = preprocess_and_balance(train_df, test_df, features, target)

    X_train, y_train = create_sequences(train_balanced, features, target, seq_len)
    X_test, y_test = create_sequences(test_processed, features, target, seq_len)

    print("TFT Shapes:", X_train.shape, y_train.shape)

    model = build_tft_model(seq_len, len(features))

    cbs = [
        EarlyStopping(patience=12, restore_best_weights=True),
        ReduceLROnPlateau(patience=5, factor=0.5)
    ]

    history = model.fit(
        X_train, y_train,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        validation_split=0.2,
        callbacks=cbs,
        verbose=1
    )

    plot_history(history, "TFT Loss")

    y_prob = model.predict(X_test).ravel()
    evaluate_and_print(y_test, y_prob)

    return model, scaler, test_processed, features



# -----------------------------
# Model C: TabNet + Sequence hybrid (custom sparsemax implemented)
# -----------------------------
def sparsemax(logits, axis=-1):
    logits = tf.cast(logits, tf.float32)
    z = tf.sort(logits, direction='DESCENDING', axis=axis)
    z_cumsum = tf.cumsum(z, axis=axis)
    k = tf.range(1, tf.shape(z)[-1] + 1, dtype=tf.float32)
    k_shape = [1] * (len(logits.shape) - 1) + [-1]
    k = tf.reshape(k, k_shape)
    support = tf.cast(1 + k * z > z_cumsum, tf.float32)
    k_z = tf.reduce_sum(support, axis=axis, keepdims=True)
    tau = (tf.reduce_sum(z * support, axis=axis, keepdims=True) - 1.0) / k_z
    return tf.maximum(0.0, logits - tau)

def tabnet_block(inputs, input_dim, feature_dim=64, output_dim=32, steps=4, gamma=1.5):

    # Prior initialized as ones but using a Lambda layer
    prior = Lambda(lambda x: tf.ones_like(x))(inputs)

    outputs = []

    for i in range(steps):

        # Step 1: Attentive transformer (Mask)
        mask_logits = Dense(input_dim)(prior)
        
        # sparsemax using Lambda so Keras accepts it
        mask = Lambda(lambda x: sparsemax(x))(mask_logits)

        # masked inputs
        masked_data = Multiply()([inputs, mask])

        # Step 2: Feature transformer
        feat = Dense(feature_dim, activation="relu")(masked_data)
        feat = Dense(output_dim, activation="relu")(feat)

        outputs.append(feat)

        # Update prior = prior * (gamma - mask)
        prior = Multiply()([prior, Lambda(lambda x: gamma - x)(mask)])

    # Combine outputs
    out = Add()(outputs)  
    return out


def sequence_branch(seq_inp, hidden=64):
    # small CNN+GRU branch
    x = Conv1D(32, 3, padding='same', activation='relu')(seq_inp)
    x = MaxPooling1D(2)(x)
    x = GRU(hidden, return_sequences=False)(x)
    x = Dense(64, activation='relu')(x)
    return x

def build_tabnet_hybrid(seq_len, n_features, tab_feature_dim=64, tab_output_dim=32, seq_hidden=64):
    seq_inp = Input(shape=(seq_len, n_features))
    tab_inp = Input(shape=(n_features,))

    seq_out = sequence_branch(seq_inp, hidden=seq_hidden)
    tab_out = tabnet_block(tab_inp, input_dim=n_features, feature_dim=tab_feature_dim, output_dim=tab_output_dim, steps=4)

    merged = Concatenate()([seq_out, tab_out])
    x = Dense(64, activation='relu')(merged)
    x = Dropout(0.3)(x)
    out = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=[seq_inp, tab_inp], outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                  loss='binary_crossentropy',
                  metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), tf.keras.metrics.AUC()])
    model.summary()
    return model

# -----------------------------
# Training + evaluation helpers
# -----------------------------
def plot_history(history, title="Training"):
    plt.figure(figsize=(10,4))
    plt.plot(history.history['loss'], label='train loss')
    if 'val_loss' in history.history:
        plt.plot(history.history['val_loss'], label='val loss')
    plt.title(title)
    plt.legend()
    plt.show()

def evaluate_and_print(y_true, y_prob):
    y_pred = (y_prob > 0.5).astype(int).ravel()
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Precision:", precision_score(y_true, y_pred))
    print("Recall:", recall_score(y_true, y_pred))
    print("F1:", f1_score(y_true, y_pred))
    print("ROC-AUC:", roc_auc_score(y_true, y_prob))
    print("Confusion matrix:\n", confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred, digits=4))

# -----------------------------
# End-to-end train functions
# -----------------------------
def train_cnn_lstm_pipeline(seq_len=SEQ_LENGTH):
    train_df, test_df, features, target = load_and_prepare_data()
    train_balanced, test_processed, scaler = preprocess_and_balance(train_df, test_df, features, target)

    X_train, y_train = create_sequences(train_balanced, features, target, seq_len)
    X_test, y_test = create_sequences(test_processed, features, target, seq_len)

    print("Shapes:", X_train.shape, y_train.shape, X_test.shape, y_test.shape)
    model = build_cnn_lstm(seq_len, len(features))
    cbs = [EarlyStopping(patience=12, restore_best_weights=True), ReduceLROnPlateau(patience=5, factor=0.5)]
    history = model.fit(X_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, validation_split=0.2, callbacks=cbs, verbose=1)
    plot_history(history, "CNN-LSTM Loss")
    y_prob = model.predict(X_test).ravel()
    evaluate_and_print(y_test, y_prob)
    return model, scaler, test_processed, features

def train_transformer_pipeline(seq_len=SEQ_LENGTH):
    train_df, test_df, features, target = load_and_prepare_data()
    train_balanced, test_processed, scaler = preprocess_and_balance(train_df, test_df, features, target)

    X_train, y_train = create_sequences(train_balanced, features, target, seq_len)
    X_test, y_test = create_sequences(test_processed, features, target, seq_len)

    model = build_transformer_binary(seq_len, len(features))
    cbs = [EarlyStopping(patience=12, restore_best_weights=True), ReduceLROnPlateau(patience=5, factor=0.5)]
    history = model.fit(X_train, y_train, epochs=EPOCHS, batch_size=BATCH_SIZE, validation_split=0.2, callbacks=cbs, verbose=1)
    plot_history(history, "Transformer Loss")
    y_prob = model.predict(X_test).ravel()
    evaluate_and_print(y_test, y_prob)
    return model, scaler, test_processed, features

def train_tabnet_hybrid_pipeline(seq_len=SEQ_LENGTH):
    train_df, test_df, features, target = load_and_prepare_data()
    train_balanced, test_processed, scaler = preprocess_and_balance(train_df, test_df, features, target)

    X_train_seq, y_train = create_sequences(train_balanced, features, target, seq_len)
    X_test_seq, y_test = create_sequences(test_processed, features, target, seq_len)

    # For tabular branch we use the last day's features as tabular input
    X_train_tab = X_train_seq[:, -1, :]
    X_test_tab = X_test_seq[:, -1, :]

    model = build_tabnet_hybrid(seq_len, len(features))
    # compute class weights from original train (before SMOTE)
    orig_train_df = pd.concat([train_df for train_df in [train_df]], ignore_index=False)  # placeholder to keep same var name
    # compute class weights from y_train (balanced training after SMOTE not appropriate for weighting)
    cw_vals = compute_class_weight("balanced", classes=np.array([0,1]), y=y_train)
    class_weights = {0: cw_vals[0], 1: cw_vals[1]}

    cbs = [EarlyStopping(patience=12, restore_best_weights=True), ReduceLROnPlateau(patience=5, factor=0.5)]
    history = model.fit([X_train_seq, X_train_tab], y_train, epochs=EPOCHS, batch_size=BATCH_SIZE,
                        validation_split=0.2, callbacks=cbs, class_weight=class_weights, verbose=1)
    plot_history(history, "TabNet+Seq Loss")
    y_prob = model.predict([X_test_seq, X_test_tab]).ravel()
    evaluate_and_print(y_test, y_prob)
    return model, scaler, test_processed, features

# -----------------------------
# Main: choose which pipelines to run
# -----------------------------
if __name__ == "__main__":
    # Pick the model(s) you want to run. Running all three will take longer.
    print("Running CNN-LSTM pipeline...")
    model_cnn, scaler, test_processed, features = train_cnn_lstm_pipeline(seq_len=SEQ_LENGTH)

    print("\nRunning Transformer pipeline...")
    model_tr, scaler, test_processed, features = train_transformer_pipeline(seq_len=SEQ_LENGTH)

    print("\nRunning TFT pipeline...")
    model_tft, scaler, test_processed, features = train_tft_pipeline(seq_len=SEQ_LENGTH)
```
